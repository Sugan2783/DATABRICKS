```python
l = [2,4,5,6]
l1 = [1,2,4,5,7]

df1 = spark.createDataFrame(l, ["id"])
df2 = spark.createDataFrame(l1, ["id"])
df1.show()  
df2.show()


df1.exceptAll(df2).show()

df_join = df1.join(df2, on ="id", how ="leftanti").show()


df_join_semi = df1.join(df2, on ="id", how ="leftsemi").show()
```

```python
ids = [a for a in df2.select("id").collect()]
print(ids)

ids = [a.id for a in df2.select("id").collect()]
print(ids)
```

```python
ids = [a.id for a in df2.select("id").collect()]
print(ids)

df1.filter(df1.id.isin(ids)).show()
```

```python
df1.subtract(df2).show()

df1.subtract(df1.join(df2, on ="id", how ="inner")).show()
```

```python
l = [2,4,5,6]
l1 = [(1,'A'),(2,'B'),(4,'C'),(5,'D'),(7,'E')]

df1 = spark.createDataFrame(l, ["id"])
df2 = spark.createDataFrame(l1, ["id", "name"])
df1.show()  
df2.show()
   
```

```python
df1.exceptAll(df2).show()

df_join = df1.join(df2, on ="id", how ="leftanti").show()
```

```python
# Select only the 'id' column from both DataFrames before exceptAll
df1_id = df1.select("id")
df2_id = df2.select("id")
display(
    df1_id.exceptAll(df2_id)
)

# For leftanti join, assign the result to a variable and then display
df_join = df1.join(
    df2,
    on="id",
    how="leftanti"
)
display(df_join)
```
