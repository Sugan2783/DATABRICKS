```python
%sql
-- Create the table
CREATE TABLE ElementTable (
    Element CHAR(1)
);

-- Insert values
INSERT INTO ElementTable (Element) VALUES ('A');
INSERT INTO ElementTable (Element) VALUES ('B');
INSERT INTO ElementTable (Element) VALUES ('C');
INSERT INTO ElementTable (Element) VALUES ('D');
INSERT INTO ElementTable (Element) VALUES ('E');
INSERT INTO ElementTable (Element) VALUES ('F');
```

```python
%sql
select * from ElementTable;
```

```python
%sql
with cte1 as
(
  select element, 1 as level from ElementTable
  union all
  select element, level + 1 as seq from cte1
  where level < length(element)
)
select element, level as seq from cte1;
```

```python
%sql
select
  element,
  seq as level
from (
  select
    element,
    explode(
      sequence(1, length(element))
    ) as seq
  from ElementTable
)
```
