```python
from pyspark.sql.functions import year, month, dayofmonth, dayofweek, quarter, date_format

df = spark.sql('''select explode(sequence(to_date('2025-01-01'),
                        to_date('2025-02-01'), interval 1 day)) as date''')

df = df.withColumn('year', year('date')).withColumn('month',month('date')).withColumn('day', dayofmonth('date')).withColumn('dayofweek', dayofweek('date')).withColumn('quarter', quarter('date')).withColumn('date', date_format('date', 'yyyy-MM-dd'))
df.display()
```

```python
df = df.withColumn('day_name', date_format('date', 'EEEE')).withColumn('monthname', date_format('date', 'MMMM'))
df.display()
```

```python
def is_valid_email(email):
    if '@' in email and '.' in email:
        print('yes valid email')
    print('no valid email')

def is_valid(email):
    username, domain = email.split('@')
    return {"username" : username,
            "domain" : domain}
            

is_valid_email('johndoe@gmailcom')

is_valid("john.doe@gmail.com")
```

```python
def write_log(message):
    with open('log.txt', 'a') as f:
        f.write(message + '\n')
    print(message)
```

```python
email = input("Please Enter Your Email: ")

print(is_valid(email))

if not is_valid(email):
    write_log(f" Invalid EmIL RECIEVED: {email}")
else:
    is_valid(email)
```

```python
from pyspark.sql.functions import col, udf
from pyspark.sql.types import StringType

def clean_name(name):
    return name.strip().lower().capitalize()

clean_name_udf = udf(clean_name, StringType())

df = spark.createDataFrame([
    (' MARisa ',),
    ('  Dan ',),
    ('  Bob  ',)
], ['name'])
df.select(clean_name_udf(col('name')).alias('Name')).show()
```
