```python
df = spark.sql(f"""
SELECT 
  table_catalog,
  table_schema,
  table_name,
  table_type,
  table_owner,
  created,
  last_altered,
  comment
FROM information_schema.tables
WHERE table_catalog = 'bigdata'
  AND table_schema = 'customer_dlt'
""")
df.display()
```
