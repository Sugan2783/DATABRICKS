```python
def top_product_rate(df, col_top_rate_product, col_rate, fil_val):

    from pyspark.sql.functions import count, desc, lit, col

    top_rate = df.groupBy(col_top_rate_product)\
        .agg(count(col_top_rate_product).alias("top_count"))\
        .orderBy(desc("top_count"))

    top_prd_fil = df.filter(col(col_rate) == fil_val)

    cnt_prod_feed = df.count()
    

    return top_rate, top_prd_fil, cnt_prod_feed

top_rate, top_prd_fil, cnt_prod_feed = top_product_rate(df_cust, 'Product', 'Comment', 'Excellent')
# top_rate, top_prd_fil, cnt_prod_feed = top_product_rate(df_cust, df_cust.columns[1], df_cust.columns[3], 'Excellent')

top_rate.display()
top_prd_fil.display()
print(cnt_prod_feed)
```
