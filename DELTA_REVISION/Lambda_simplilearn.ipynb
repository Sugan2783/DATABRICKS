```python
num = [10,30,49,67,89,90]
greater = list(filter(lambda num : num > 30, num))
print(greater)
```

```python
num = [10, 30, 49, 67, 89, 90]
greater = [n for n in num if n > 30]
display(greater)
```

```python
cube = map(lambda num : pow(num, 3), num)
print(list(cube))
```

```python
def quadratic(a,b,c):
    return lambda x: a*x**2 + b*2 + c
f = quadratic(2,3,5)
print(f(2))
```

```python
#find the avg val from list of numbers

list_num = [10,30,40,50,60]
sum = 0
for i in list_num:
    sum = sum + i
    result = sum / len(list_num)
print(sum)
print(result)
```

```python
#break

for i in range(1,10):
    if i == 5:
        break
    print(i)

#continue
for i in range(1,10+1):
    if i == 5:
        continue
    print(i)
```
