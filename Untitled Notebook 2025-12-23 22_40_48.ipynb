```python
import pandas as pd

df_claims = pd.read_csv("/Volumes/prod/bronze/projectdata/claims.csv")

print(df_claims.columns)

df_cl = spark.createDataFrame(df_claims)

# df_cl = df_claims.toDF(df_claims)

display(df_cl)
```

```python
from pyspark.sql.functions import col, sum

def null_checks(df, null_check_cols):
    agg_exprs = [
        sum(
            col(col_name).isNull().cast("int")
        ).alias("null_cnt_" + col_name)
        for col_name in null_check_cols
    ]
    return df.agg(*agg_exprs)

null_val = null_checks(df_cl, df_cl.columns)
display(null_val)
```
